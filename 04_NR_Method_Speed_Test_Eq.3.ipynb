```julia
function Newton_Raphson_Classic(Bo, B1, B2, B3, B4, λo, epsd)
    
         EPS=epsd
         LoPo=0
         ITNumb=0
         Jack=0
         
    ####################################################
    #                                                ###
    # THE METHOD FIRST PART                          ###     
         
         λo=λo                                       
         f=Bo+B1*λo+B2*λo^2+B3*λo^3+B4*λo^4          
         fprime=B1+2*B2*λo+3*B3*λo^2+4*B4*λo^3       
         u=f/fprime                                  
         w=λo-u                                      
    #                                                ###
    ####################################################
    #                               ####################
    # THE CHECKER FIRST             ####################             
                                    ####################
         if isnan(w)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(w)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(w)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1                     ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=λo                             ##############
         ITNumb=ITNumb+0                  ##############
         eps=0                            ##############
         Eps=0                            ##############
            else                          ##############  
         λ=w                              ##############
         eps=abs((λ-λo)/λ)                ##############
         Eps=eps                          ##############
         ITNumb=ITNumb+1                  ##############
            end                           ##############
                                          ##############
  ######################################################
  #                       ##############################
  # THE LOOP OPENER       ##############################   
  #                       ##############################
        while eps>EPS     ##############################      
        if ITNumb>1000    ##############################
           eps=0          ##############################
           Jack=Jack+1    ############################## 
       else               ##############################
  ######################################################
  # #                                                ###
  # # THE METHOD FINAL PART                          ###     
  # #                                                ###                                       
         λo=λ
         f=Bo+B1*λo+B2*λo^2+B3*λo^3+B4*λo^4          
         fprime=B1+2*B2*λo+3*B3*λo^2+4*B4*λo^3       
         u=f/fprime                                  
         w=λo-u                                      
  # #                                                ###
  ######################################################
  # #                               ####################
  # # THE CHECKER SECOND            ####################             
                                    ####################
         if isnan(w)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(w)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(w)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################          
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER FINAL                 ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=λo                             ##############
         ITNumb=ITNumb+0                  ##############
         eps=0                            ##############
            else                          ##############  
         λ=w                              ##############
         eps=abs((λ-λo)/λ)                ##############
         Eps=eps                          ##############       
         ITNumb=ITNumb+1                  ##############
            end                           ############## 
            end                           ##############
        end                               ##############
  # #                                     ##############
  ######################################################
 λD=λ   
 ITNumbD=ITNumb
 res=[λD, ITNumbD]
 return(res)
 end
```

```julia
function Super16_4P16R(Ao, A1, A2, A3, A4, λo, epsd)# αd in degrees #####################   

EPS=epsd    
LoPo=0
ITNumb=0
Jack=0
    
 ###########################################################
    #                                                       ###
    # THE METHOD PART 1.1.                                  ###   
    
         λo=λo                                              
         f=Ao+A1*λo+A2*λo^2+A3*λo^3+A4*λo^4        
         fprime=A1+2*A2*λo+3*A3*λo^2+4*A4*λo^3     
         u=f/fprime                                
         w=λo-u                                    
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.1.              ####################             
                                    ####################
         if isnan(w)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(w)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(w)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.1                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=λo                             ##############
         ITNumb=ITNumb+0                  ##############
         eps=0                            ##############
         Eps=0                            ##############
            else                          ##############  
                                          ##############
    ###########################################################
    #                                                       ###
    # THE METHOD PART 1.2.                                  ###   
    
         fw=Ao+A1*w+A2*w^2+A3*w^3+A4*w^4
         z=λo-u*((f-fw)/(f-2*fw))                              
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.2.              ####################             
                                    ####################
         if isnan(z)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(z)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(z)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.2                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=w                              ##############
         ITNumb=ITNumb+0.25               ##############
         eps=0                            ##############
         Eps=abs((λ-λo)/λ)                ##############
            else                          ##############  
                                          ##############
    ###########################################################
    #                                                       ###
    # THE METHOD PART 1.3.                                  ###   
    
         fz=Ao+A1*z+A2*z^2+A3*z^3+A4*z^4
         fwx=(fw-f)/(w-λo)
         fwxx=(fwx-fprime)/(w-λo) 
         fzx=(fz-f)/(z-λo)
         fzxx=(fzx-fprime)/(z-λo)    
         a1=fprime
         a2=(fwxx*(z-λo)-fzxx*(w-λo))/(z-w)   
         a3=(fzxx-fwxx)/(z-w)
         qprimez=a1+2*a2*(z-λo)+3*a3*(z-λo)^2   
         y=z-fz/qprimez   
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.3.              ####################             
                                    ####################
         if isnan(y)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(y)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(y)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.3                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=z                              ##############
         ITNumb=ITNumb+0.50               ##############
         eps=0                            ##############
         Eps=abs((λ-λo)/λ)                ##############
            else                          ##############  
                                          ##############
  ###########################################################
  #                                                       ###
  # THE METHOD PART 1.4.                                  ###   
    
         fy=Ao+A1*y+A2*y^2+A3*y^3+A4*y^4
         fwx=(fw-f)/(w-λo)
         fwxx=(fwx-fprime)/(w-λo) 
         fzx=(fz-f)/(z-λo)
         fzxx=(fzx-fprime)/(z-λo) 
         fyx=(fy-f)/(y-λo)       
         fyxx=(fyx-fprime)/(y-λo) 
         S1=w-λo 
         S2=z-λo
         S3=y-λo    
         b1=fprime
         b2=(fwxx*(-S2*S2*S3+S2*S3*S3)+fzxx*(S1*S1*S3-S1*S3*S3)+fyxx*(-S1*S1*S2+S1*S2*S2))/(-S1*S1*S2+S1*S2*S2+S1*S1*S3-S2*S2*S3-S1*S3*S3+S2*S3*S3)   
         b3=(fwxx*(S2*S2-S3*S3)+fzxx*(-S1*S1+S3*S3)+fyxx*(S1*S1-S2*S2))/(-S1*S1*S2+S1*S2*S2+S1*S1*S3-S2*S2*S3-S1*S3*S3+S2*S3*S3) 
         b4=(fwxx*(-S2+S3)+fzxx*(S1-S3)+fyxx*(-S1+S2))/(-S1*S1*S2+S1*S2*S2+S1*S1*S3-S2*S2*S3-S1*S3*S3+S2*S3*S3)       
         rprimey=b1+2*b2*(y-λo)+3*b3*(y-λo)^2+4*b4*(y-λo)^3   
         l=y-fy/rprimey              
  ###########################################################    
    #                               ####################
    # THE CHECKER 1.4.              ####################             
                                    ####################
         if isnan(l)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(l)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(l)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.4                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=y                              ##############
         ITNumb=ITNumb+0.75               ##############
         eps=0                            ##############
         Eps=abs((λ-λo)/λ)                ##############
            else                          ##############  
                                          ##############
 #######################################################
 # THE ITERATION FINISHER                     ##########
 #                                            ##########     
             λ=l                              ##########
             ITNumb=ITNumb+1                  ##########
             eps=abs((λ-λo)/λ)                ##########
             Eps=eps                          ##########
                end                           ##########
            end                               ##########
            end                               ##########
        end                                   ##########
  #                                           ##########
  ######################################################  
  ######################################################              
                
  ######################################################
  #                       ##############################
  # THE LOOP OPENER       ##############################   
  #                       ##############################
        while eps>EPS     ##############################      
        if ITNumb>1000    ##############################
           eps=0          ##############################
           Jack=Jack+1    ############################## 
       else               ##############################
  ###################################################### 
     ###########################################################
    #                                                       ###
    # THE METHOD PART 1.1.                                  ###   
    
         λo=λo                                              
         f=Ao+A1*λo+A2*λo^2+A3*λo^3+A4*λo^4        
         fprime=A1+2*A2*λo+3*A3*λo^2+4*A4*λo^3     
         u=f/fprime                                
         w=λo-u                                    
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.1.              ####################             
                                    ####################
         if isnan(w)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(w)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(w)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.1                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=λo                             ##############
         ITNumb=ITNumb+0                  ##############
         eps=0                            ##############
         Eps=0                            ##############
            else                          ##############  
                                          ##############
    ###########################################################
    #                                                       ###
    # THE METHOD PART 1.2.                                  ###   
    
         fw=Ao+A1*w+A2*w^2+A3*w^3+A4*w^4
         z=λo-u*((f-fw)/(f-2*fw))                              
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.2.              ####################             
                                    ####################
         if isnan(z)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(z)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(z)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.2                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=w                              ##############
         ITNumb=ITNumb+0.25               ##############
         eps=0                            ##############
         Eps=abs((λ-λo)/λ)                ##############
            else                          ##############  
                                          ##############
    ###########################################################
    #                                                       ###
    # THE METHOD PART 1.3.                                  ###   
    
         fz=Ao+A1*z+A2*z^2+A3*z^3+A4*z^4
         fwx=(fw-f)/(w-λo)
         fwxx=(fwx-fprime)/(w-λo) 
         fzx=(fz-f)/(z-λo)
         fzxx=(fzx-fprime)/(z-λo)    
         a1=fprime
         a2=(fwxx*(z-λo)-fzxx*(w-λo))/(z-w)   
         a3=(fzxx-fwxx)/(z-w)
         qprimez=a1+2*a2*(z-λo)+3*a3*(z-λo)^2   
         y=z-fz/qprimez   
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.3.              ####################             
                                    ####################
         if isnan(y)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(y)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(y)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.3                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=z                              ##############
         ITNumb=ITNumb+0.50               ##############
         eps=0                            ##############
         Eps=abs((λ-λo)/λ)                ##############
            else                          ##############  
                                          ##############
  ###########################################################
  #                                                       ###
  # THE METHOD PART 1.4.                                  ###   
    
         fy=Ao+A1*y+A2*y^2+A3*y^3+A4*y^4
         fwx=(fw-f)/(w-λo)
         fwxx=(fwx-fprime)/(w-λo) 
         fzx=(fz-f)/(z-λo)
         fzxx=(fzx-fprime)/(z-λo) 
         fyx=(fy-f)/(y-λo)       
         fyxx=(fyx-fprime)/(y-λo) 
         S1=w-λo 
         S2=z-λo
         S3=y-λo    
         b1=fprime
         b2=(fwxx*(-S2*S2*S3+S2*S3*S3)+fzxx*(S1*S1*S3-S1*S3*S3)+fyxx*(-S1*S1*S2+S1*S2*S2))/(-S1*S1*S2+S1*S2*S2+S1*S1*S3-S2*S2*S3-S1*S3*S3+S2*S3*S3)   
         b3=(fwxx*(S2*S2-S3*S3)+fzxx*(-S1*S1+S3*S3)+fyxx*(S1*S1-S2*S2))/(-S1*S1*S2+S1*S2*S2+S1*S1*S3-S2*S2*S3-S1*S3*S3+S2*S3*S3) 
         b4=(fwxx*(-S2+S3)+fzxx*(S1-S3)+fyxx*(-S1+S2))/(-S1*S1*S2+S1*S2*S2+S1*S1*S3-S2*S2*S3-S1*S3*S3+S2*S3*S3)       
         rprimey=b1+2*b2*(y-λo)+3*b3*(y-λo)^2+4*b4*(y-λo)^3   
         l=y-fy/rprimey              
  ###########################################################    
    #                               ####################
    # THE CHECKER 1.4.              ####################             
                                    ####################
         if isnan(l)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(l)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(l)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.4                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=y                              ##############
         ITNumb=ITNumb+0.75               ##############
         eps=0                            ##############
         Eps=abs((λ-λo)/λ)                ##############
            else                          ##############  
                                          ##############
 #######################################################
 # THE ITERATION FINISHER                     ##########
 #                                            ##########     
             λ=l                              ##########
             ITNumb=ITNumb+1                  ##########
             eps=abs((λ-λo)/λ)                ##########
             Eps=eps                          ##########
                end                           ##########
            end                               ##########
            end                               ##########
        end                                   ##########
  #                                           ##########
  ###################################################### 
                                              ########## 
            end # If ITNumb>1000 Loop Close   ##########  
        end # While Loop Close                ##########        
  #                                           ##########
  ######################################################  
  ######################################################
 λD=λ   
 ITNumbD=ITNumb
 res=[λD, ITNumbD]
 return(res)
 end
```
