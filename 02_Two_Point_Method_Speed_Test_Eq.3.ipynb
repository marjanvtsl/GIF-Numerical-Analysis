```julia
function Ostrowski_2P4R(Bo, B1, B2, B3, B4, λo, EPSD)# αd in degrees #####################  

EPS=EPSD  
nano50=0    
nan=0
nan10=0
info50=0    
inf=0
inf10=0
zeroo50=0
zero=0
zero10=0
LoPo=0
ITNumb=0
Jack=0

    
    ####################################################
    #                                                ###
    # THE METHOD PART 1.1                            ###     
         
         λo=λo                                       
         f=Bo+B1*λo+B2*λo^2+B3*λo^3+B4*λo^4          
         fprime=B1+2*B2*λo+3*B3*λo^2+4*B4*λo^3
         u=f/fprime                               
         w=λo-u                                      
    #                                                ###
    ####################################################
    #                               ####################
    # THE CHECKER 1.1               ####################             
                                    ####################
         if isnan(w)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(w)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(w)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
         if LoPo==1                 ####################
         nano50=nano50+1            ####################
         elseif LoPo==2             ####################
         info50=info50+1            ####################
         elseif LoPo==3             ####################
         zeroo50=zeroo50+1          ####################
                        end         ####################
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.1                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=λo                             ##############
         ITNumb=ITNumb+0                  ##############
         eps=0                            ##############
         Eps=0                            ##############
            else                          ##############  
                                          ##############
    ###########################################################
    #                                                       ###
    # THE METHOD PART 1.2.                                  ###   
    
         fw=Bo+B1*w+B2*w^2+B3*w^3+B4*w^4
         z=w-(fw/fprime)*f/(f-2*fw)    
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.2.              ####################             
                                    ####################
         if isnan(z)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(z)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(z)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
         if LoPo==1                 ####################
         nan10=nan10+1              ####################
         elseif LoPo==2             ####################
         inf10=inf10+1              ####################
         elseif LoPo==3             ####################
         zero10=zero10+1            ####################
                        end         ####################
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.2                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=w                              ##############
         ITNumb=ITNumb+0.5                ##############
         eps=0                            ##############
         Eps=abs((λ-λo)/λ)                ##############
            else                          ##############  
                                          ##############
 #######################################################
 # THE ITERATION FINISHER                     ##########
 #                                            ##########     
             λ=z                              ##########
             ITNumb=ITNumb+1                  ##########
             eps=abs((λ-λo)/λ)                ##########
             Eps=eps                          ##########
                end                           ##########
            end                               ##########
  #                                           ##########
  ######################################################  
  L1=λ                                        #  #  #  #
  Eps1=Eps                                    #  #  #  #
  ######################################################            
  #                       ##############################
  # THE LOOP OPENER       ##############################   
  #                       ##############################
        while eps>EPS     ##############################      
        if ITNumb>1000    ##############################
           eps=0          ##############################
           Jack=Jack+1    ############################## 
       else               ##############################
  ######################################################

    ####################################################
    #                                                ###
    # THE METHOD PART 1.1                            ###     
         
         λo=λ                                       
         f=Bo+B1*λo+B2*λo^2+B3*λo^3+B4*λo^4          
         fprime=B1+2*B2*λo+3*B3*λo^2+4*B4*λo^3
         u=f/fprime                               
         w=λo-u                                      
    #                                                ###
    ####################################################
    #                               ####################
    # THE CHECKER 1.1               ####################             
                                    ####################
         if isnan(w)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(w)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(w)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
         if LoPo==1                 ####################
         nan=nan+1                  ####################
         elseif LoPo==2             ####################
         inf=inf+1                  ####################
         elseif LoPo==3             ####################
         zero=zero+1                ####################
                        end         ####################
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.1                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=λo                             ##############
         ITNumb=ITNumb+0                  ##############
         eps=0                            ##############
         Eps=0                            ##############
            else                          ##############  
                                          ##############
    ###########################################################
    #                                                       ###
    # THE METHOD PART 1.2.                                  ###   
    
         fw=Bo+B1*w+B2*w^2+B3*w^3+B4*w^4
         z=w-(fw/fprime)*f/(f-2*fw)     
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.2.              ####################             
                                    ####################
         if isnan(z)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(z)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(z)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
         if LoPo==1                 ####################
         nan=nan+1                  ####################
         elseif LoPo==2             ####################
         inf=inf+1                  ####################
         elseif LoPo==3             ####################
         zero=zero+1                ####################
                        end         ####################
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.2                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=w                              ##############
         ITNumb=ITNumb+0.5                ##############
         eps=0                            ##############
         Eps=abs((λ-λo)/λ)                ##############
            else                          ##############  
                                          ##############
 #######################################################
 # THE ITERATION FINISHER                     ##########
 #                                            ##########     
             λ=z                              ##########
             ITNumb=ITNumb+1                  ##########
             eps=abs((λ-λo)/λ)                ##########
             Eps=eps                          ##########
                end                           ##########
            end                               ##########
            end # If ITNumb>1000 Loop Close   ##########  
        end # While Loop Close                ##########        
  #                                           ##########
  ######################################################                        
 λD=λ   
 ITNumbD=ITNumb
 res=[λD, ITNumbD]
 return(res)
 end
```
