```julia
function ShSh_3P8R(Ao, A1, A2, A3, A4, λo, EPSD)# αd in degrees #####################  

EPS=EPSD  
nano33=0
info33=0
zeroo33=0
nano67=0
info67=0
zeroo67=0
nan10=0
inf10=0
zero10=0
nan=0
inf=0
zero=0    
LoPo=0
ITNumb=0
Jack=0  
 ###########################################################
    #                                                       ###
    # THE METHOD PART 1.1.                                  ###   
    
         λo=λo                                              
         f=Ao+A1*λo+A2*λo^2+A3*λo^3+A4*λo^4        
         fprime=A1+2*A2*λo+3*A3*λo^2+4*A4*λo^3     
         u=f/fprime                                
         w=λo-u                                    
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.1.              ####################             
                                    ####################
         if isnan(w)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(w)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(w)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
         if LoPo==1                 ####################
         nano33=nano33+1            ####################
         elseif LoPo==2             ####################
         info33=info33+1            ####################
         elseif LoPo==3             ####################
         zeroo33=zeroo33+1          ####################
                        end         ####################
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.1                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=λo                             ##############
         ITNumb=ITNumb+0                  ##############
         eps=0                            ##############
         Eps=0                            ##############
            else                          ##############  
                                          ##############
    ###########################################################
    #                                                       ###
    # THE METHOD PART 1.2.                                  ###   
    
         fw=Ao+A1*w+A2*w^2+A3*w^3+A4*w^4
         z=w-u*fw/(f-2*fw)                              
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.2.              ####################             
                                    ####################
         if isnan(z)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(z)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(z)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
         if LoPo==1                 ####################
         nano67=nano67+1            ####################
         elseif LoPo==2             ####################
         info67=info67+1            ####################
         elseif LoPo==3             ####################
         zeroo67=zeroo67+1          ####################
                        end         ####################
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.2                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=w                              ##############
         ITNumb=ITNumb+0.33               ##############
         eps=0                            ##############
         Eps=abs((λ-λo)/λ)                ##############
            else                          ##############  
                                          ##############
    ###########################################################
    #                                                       ###
    # THE METHOD PART 1.3.                                  ###   
    
         fz=Ao+A1*z+A2*z^2+A3*z^3+A4*z^4
         m=fz/f
         W=1+m+m^2      
         fxw=(fw-f)/(w-λo)
         fxz=(fz-f)/(z-λo)
         fwz=(fz-fw)/(z-w)
         l=z-W*((fxw*fz)/(fxz*fwz)) 
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.3.              ####################             
                                    ####################
         if isnan(l)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(l)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(l)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
         if LoPo==1                 ####################
         nan10=nan10+1              ####################
         elseif LoPo==2             ####################
         inf10=inf10+1              ####################
         elseif LoPo==3             ####################
         zero10=zero10+1            ####################
                        end         ####################
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.3                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=z                              ##############
         ITNumb=ITNumb+0.67               ##############
         eps=0                            ##############
         Eps=abs((λ-λo)/λ)                ##############
            else                          ##############  
                                          ##############
 #######################################################
 # THE ITERATION FINISHER                     ##########
 #                                            ##########     
             λ=l                              ##########
             ITNumb=ITNumb+1                  ##########
             eps=abs((λ-λo)/λ)                ##########
             Eps=eps                          ##########
                end                           ##########
            end                               ##########
        end                                   ##########
  #                                           ##########
  ######################################################  
  ######################################################
  #                       ##############################
  # THE LOOP OPENER       ##############################   
  #                       ##############################
        while eps>EPS     ##############################      
        if ITNumb>1000    ##############################
           eps=0          ##############################
           Jack=Jack+1    ############################## 
       else               ##############################
  ###################################################### 
    ###########################################################
    #                                                       ###
    # THE METHOD PART 1.1.                                  ###   
    
         λo=λ                                             
         f=Ao+A1*λo+A2*λo^2+A3*λo^3+A4*λo^4        
         fprime=A1+2*A2*λo+3*A3*λo^2+4*A4*λo^3     
         u=f/fprime                                
         w=λo-u                           
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.1.              ####################             
                                    ####################
         if isnan(w)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(w)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(w)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
         if LoPo==1                 ####################
         nan=nan+1                  ####################
         elseif LoPo==2             ####################
         inf=inf+1                  ####################
         elseif LoPo==3             ####################
         zero=zero+1                ####################
                        end         ####################
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.1                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=λo                             ##############
         ITNumb=ITNumb+0                  ##############
         eps=0                            ##############
         Eps=Eps                          ##############
            else                          ##############  
                                          ##############
    ###########################################################
    #                                                       ###
    # THE METHOD PART 1.2.                                  ###   
    
         fw=Ao+A1*w+A2*w^2+A3*w^3+A4*w^4
         z=w-u*fw/(f-2*fw)                              
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.2.              ####################             
                                    ####################
         if isnan(z)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(z)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(z)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
         if LoPo==1                 ####################
         nan=nan+1                  ####################
         elseif LoPo==2             ####################
         inf=inf+1                  ####################
         elseif LoPo==3             ####################
         zero=zero+1                ####################
                        end         ####################
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.2                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=w                              ##############
         ITNumb=ITNumb+0.33               ##############
         eps=0                            ##############
         Eps=abs((λ-λo)/λ)                ##############
            else                          ##############  
                                          ##############
    ###########################################################
    #                                                       ###
    # THE METHOD PART 1.3.                                  ###   
    
         fz=Ao+A1*z+A2*z^2+A3*z^3+A4*z^4
         m=fz/f
         W=1+m+m^2      
         fxw=(fw-f)/(w-λo)
         fxz=(fz-f)/(z-λo)
         fwz=(fz-fw)/(z-w)
         l=z-W*((fxw*fz)/(fxz*fwz)) 
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.3.              ####################             
                                    ####################
         if isnan(l)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(l)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(l)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
         if LoPo==1                 ####################
         nan=nan+1                  ####################
         elseif LoPo==2             ####################
         inf=inf+1                  ####################
         elseif LoPo==3             ####################
         zero=zero+1                ####################
                        end         ####################
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.3                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=z                              ##############
         ITNumb=ITNumb+0.67               ##############
         eps=0                            ##############
         Eps=abs((λ-λo)/λ)                ##############
            else                          ##############  
                                          ##############
 #######################################################
 # THE ITERATION FINISHER                     ##########
 #                                            ##########     
             λ=l                              ##########
             ITNumb=ITNumb+1                  ##########
             eps=abs((λ-λo)/λ)                ##########
             Eps=eps                          ##########
                end                           ##########
            end                               ##########
        end                                   ##########
                                              ########## 
            end # If ITNumb>1000 Loop Close   ##########  
        end # While Loop Close                ##########        
  #                                           ##########
  ######################################################  
 λD=λ   
 ITNumbD=ITNumb
 res=[λD, ITNumbD]
 return(res)
 end
```

```julia
function BWR_II_Opt_3P8R(Ao, A1, A2, A3, A4, λo, EPSD)# αd in degrees #####################  

EPS=EPSD  
nano33=0
info33=0
zeroo33=0
nano67=0
info67=0
zeroo67=0
nan10=0
inf10=0
zero10=0 
nan=0
inf=0
zero=0 
LoPo=0
ITNumb=0
Jack=0
beta=1    
 ###########################################################
    #                                                       ###
    # THE METHOD PART 1.1.                                  ###   
    
         λo=λo                                              
         f=Ao+A1*λo+A2*λo^2+A3*λo^3+A4*λo^4        
         fprime=A1+2*A2*λo+3*A3*λo^2+4*A4*λo^3     
         u=f/fprime                                
         w=λo-u                                  
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.1.              ####################             
                                    ####################
         if isnan(w)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(w)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(w)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
         if LoPo==1                 ####################
         nano33=nano33+1            ####################
         elseif LoPo==2             ####################
         info33=info33+1            ####################
         elseif LoPo==3             ####################
         zeroo33=zeroo33+1          ####################
                        end         ####################
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.1                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=λo                             ##############
         ITNumb=ITNumb+0                  ##############
         eps=0                            ##############
         Eps=0                            ##############
            else                          ##############  
                                          ##############
    ###########################################################
    #                                                       ###
    # THE METHOD PART 1.2.                                  ###   
    
         fw=Ao+A1*w+A2*w^2+A3*w^3+A4*w^4
         t=fw/f
         ht=1+4*t/(2-5*t)
         z=w-ht*fw/fprime                             
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.2.              ####################             
                                    ####################
         if isnan(z)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(z)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(z)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
         if LoPo==1                 ####################
         nano67=nano67+1            ####################
         elseif LoPo==2             ####################
         info67=info67+1            ####################
         elseif LoPo==3             ####################
         zeroo67=zeroo67+1          ####################
                        end         ####################
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.2                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=w                              ##############
         ITNumb=ITNumb+0.33               ##############
         eps=0                            ##############
         Eps=abs((λ-λo)/λ)                ##############
            else                          ##############  
                                          ##############
    ###########################################################
    #                                                       ###
    # THE METHOD PART 1.3.                                  ###   
    
         fz=Ao+A1*z+A2*z^2+A3*z^3+A4*z^4
         fzy=(fz-fw)/(z-w)
         fzxx=(z-w)*(fz-f)/(λo-z)^2   
         l=z-((f+beta*fz)/(f+(beta-2)*fz))*((fz)/(fzy+fzxx*(z-w)))
            
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.3.              ####################             
                                    ####################
         if isnan(l)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(l)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(l)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
         if LoPo==1                 ####################
         nan10=nan10+1              ####################
         elseif LoPo==2             ####################
         inf10=inf10+1              ####################
         elseif LoPo==3             ####################
         zero10=zero10+1            ####################
                        end         ####################
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.3                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=z                              ##############
         ITNumb=ITNumb+0.67               ##############
         eps=0                            ##############
         Eps=abs((λ-λo)/λ)                ##############
            else                          ##############  
                                          ##############
 #######################################################
 # THE ITERATION FINISHER                     ##########
 #                                            ##########     
             λ=l                              ##########
             ITNumb=ITNumb+1                  ##########
             eps=abs((λ-λo)/λ)                ##########
             Eps=eps                          ##########
                end                           ##########
            end                               ##########
        end                                   ##########
  #                                           ##########
  ######################################################  
  ######################################################
  #                       ##############################
  # THE LOOP OPENER       ##############################   
  #                       ##############################
        while eps>EPS     ##############################      
        if ITNumb>1000    ##############################
           eps=0          ##############################
           Jack=Jack+1    ############################## 
       else               ##############################
  ###################################################### 
    ###########################################################
    #                                                       ###
    # THE METHOD PART 1.1.                                  ###   
    
         λo=λ                                              
         f=Ao+A1*λo+A2*λo^2+A3*λo^3+A4*λo^4        
         fprime=A1+2*A2*λo+3*A3*λo^2+4*A4*λo^3     
         u=f/fprime                                
         w=λo-u
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.1.              ####################             
                                    ####################
         if isnan(w)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(w)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(w)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
         if LoPo==1                 ####################
         nan=nan+1                  ####################
         elseif LoPo==2             ####################
         inf=inf+1                  ####################
         elseif LoPo==3             ####################
         zero=zero+1                ####################
                        end         ####################
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.1                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=λo                             ##############
         ITNumb=ITNumb+0                  ##############
         eps=0                            ##############
         Eps=Eps                          ##############
            else                          ##############  
                                          ##############
    ###########################################################
    #                                                       ###
    # THE METHOD PART 1.2.                                  ###   
    
         fw=Ao+A1*w+A2*w^2+A3*w^3+A4*w^4
         t=fw/f
         ht=1+4*t/(2-5*t)
         z=w-ht*fw/fprime                              
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.2.              ####################             
                                    ####################
         if isnan(z)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(z)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(z)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
         if LoPo==1                 ####################
         nan=nan+1                  ####################
         elseif LoPo==2             ####################
         inf=inf+1                  ####################
         elseif LoPo==3             ####################
         zero=zero+1                ####################
                        end         ####################
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.2                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=w                              ##############
         ITNumb=ITNumb+0.33               ##############
         eps=0                            ##############
         Eps=abs((λ-λo)/λ)                ##############
            else                          ##############  
                                          ##############
    ###########################################################
    #                                                       ###
    # THE METHOD PART 1.3.                                  ###   
    
         fz=Ao+A1*z+A2*z^2+A3*z^3+A4*z^4
         fzy=(fz-fw)/(z-w)
         fzxx=(z-w)*(fz-f)/(λo-z)^2   
         l=z-((f+beta*fz)/(f+(beta-2)*fz))*((fz)/(fzy+fzxx*(z-w)))
    #                                                       ###
    ###########################################################    
    #                               ####################
    # THE CHECKER 1.3.              ####################             
                                    ####################
         if isnan(l)==true          ####################
         LoPo=LoPo+1                ####################
         elseif isinf(l)==true      ####################
         LoPo=LoPo+2                ####################
         elseif iszero(l)==true     ####################
         LoPo=LoPo+3                ####################
         else                       ####################
         LoPo=0                     ####################
         end                        ####################
                                    ####################      
         if LoPo==1                 ####################
         nan=nan+1                  ####################
         elseif LoPo==2             ####################
         inf=inf+1                  ####################
         elseif LoPo==3             ####################
         zero=zero+1                ####################
                        end         ####################
  #                                 ####################
  ######################################################
  #                                       ##############
  # THE DISTRIBUTER 1.3                   ##############
  #                                       ##############
            if LoPo >0                    ##############
         λ=z                              ##############
         ITNumb=ITNumb+0.67               ##############
         eps=0                            ##############
         Eps=abs((λ-λo)/λ)                ##############
            else                          ##############  
                                          ##############
 #######################################################
 # THE ITERATION FINISHER                     ##########
 #                                            ##########     
             λ=l                              ##########
             ITNumb=ITNumb+1                  ##########
             eps=abs((λ-λo)/λ)                ##########
             Eps=eps                          ##########
                end                           ##########
            end                               ##########
        end                                   ##########
                                              ########## 
            end # If ITNumb>1000 Loop Close   ##########  
        end # While Loop Close                ##########        
  #                                           ##########
  ######################################################                           
 λD=λ   
 ITNumbD=ITNumb
 res=[λD, ITNumbD]
 return(res)
 end
```
