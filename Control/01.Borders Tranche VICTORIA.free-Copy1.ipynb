```julia
function f_free(h1, dh, h2, m1, dm, m2)
     v1=[[0.0, 0.0, 0.0]]
     k1=0
     n1=0  
     alo=18/pi
     cdm=0.01
     δ1=+0.02274
     δ2=-0.6453
     δ3=+5.001
     A=1
     B=10
     n=3
     αn=0
     αm=0
     AR=15
     Jack1=0
     a=Threads.threadid()
     println("a1=$a")
for h=h1:dh:h2                        
       
                                                            #################### ISA ###############################################################
                                                            if h<11000.001
                                                               t=288.15-0.0065*h  
                                                               p=101325*(1-0.0065*h/288.15)^5.2561                
                                                               ro=1.2255*(1-0.0065*h/288.15)^4.2561
                                                            else  
                                                               t=216.65
                                                               p=22630.6*2.718281828459^(-(h-11000)/6336.09)
                                                               ro=1.60797274667397*10^(-5)*p                    
                                                            end
                                                               DIV=0.000017332*((t/288.15)^1.5)*(398.55/(t+110.4))  
                                                               kiv=DIV/ro
                                                               SOS=sqrt(1.4*287*t)
                                                               teta=t/288.15
                                                               delta=p/101325
                                                             ############### ISA END ####################################################################
############ FOR CELLS ###############    
 for Wmax=m1:dm:m2                    
     We=0.9*0.4854*Wmax^1.015
     Wm=0.5*(Wmax+We)                    
     W=[We, Wm, Wmax]                          
 for Wo in W          
     Wog=Wo*9.81
 for cc=(0.39, 1.35)                        
     fe=cc*(Wmax/1000)^(2/3)
 for bb=(0.85, 1.15)                                
     Diameter=bb*Wo^0.308
     R=Diameter/2
     DA=pi*R^2
     chord=R/AR
 for aa=(130, 150)                          
     Vt=aa*Diameter^0.171                                                                                                                                              
     omega=Vt/R
     Cw=Wog/(ro*DA*Vt^2)
     lh=sqrt(Cw/2)
     Re=0.75*Vt*chord/kiv                      
     if 10^6<Re<10^7
     cdo=0.015058-0.0010138*log10(Re)    
     else
     cdo=0.00867/(Re/2*10^6)^0.125
     end                  
     δo=cdo
     b=0.75*Vt/SOS
     c=(ro*pi*R^2*Vt^2)
 for sigma=(0.04, 0.1, 0.2)            
     s=6/sigma                      
 for V=1:1:131                               
     PDo=0.5*ro*V*V*fe
     mic=V/Vt
     a=V/SOS
     d=1.5*mic^2
     Ho=(3.65/8)*pi*ro*sigma*cdm*Vt*V*R^2
     k=0
     L1=100
     L2=100
     L3=100
     EPS=10^-4
                                                           
     for αd0 in [0, 10, 20, 30, 40, 50, 60, 70, 80, 85, 88, 89, 89.5, 89.9]
         αdo=αd0*pi/180
         Mro=(V*cos(αdo)+0.75*Vt)/SOS    
         eps=1
         ITNumb=0
         nan=0
         L=αdo
         F1=0; F2=0; F3=0; F4=0;
       
            while eps>EPS
                                           
                 if ITNumb>1000
                             Lock=round(57.3*L, digits=3)
                             Jack1=Jack1+1
                             L=100
                             eps=EPS-1
                 else                
                             αd=L 
                             f=tan(αd) - (Ho + PDo*(A + B*sin(lh^2/(mic*sqrt(mic^2*cos(αd)^2 + (mic*cos(αd)*tan(αd) + sqrt(-mic^2*cos(αd)^2/2 + sqrt(4*lh^4 + mic^4*cos(αd)^4)/2))^2)) + αd + αm)^n))/Wog 
                             fprime=-B*PDo*n*(lh^2*(mic^2*sin(αd)*cos(αd) - (mic*cos(αd)*tan(αd) + sqrt(-mic^2*cos(αd)^2/2 + sqrt(4*lh^4 + mic^4*cos(αd)^4)/2))*(2*mic*(tan(αd)^2 + 1)*cos(αd) - 2*mic*sin(αd)*tan(αd) + 2*(-mic^4*sin(αd)*cos(αd)^3/(2*sqrt(4*lh^4 + mic^4*cos(αd)^4)) + mic^2*sin(αd)*cos(αd)/2)/sqrt(-mic^2*cos(αd)^2/2 + sqrt(4*lh^4 + mic^4*cos(αd)^4)/2))/2)/(mic*(mic^2*cos(αd)^2 + (mic*cos(αd)*tan(αd) + sqrt(-mic^2*cos(αd)^2/2 + sqrt(4*lh^4 + mic^4*cos(αd)^4)/2))^2)^(3/2)) + 1)*sin(lh^2/(mic*sqrt(mic^2*cos(αd)^2 + (mic*cos(αd)*tan(αd) + sqrt(-mic^2*cos(αd)^2/2 + sqrt(4*lh^4 + mic^4*cos(αd)^4)/2))^2)) + αd + αm)^n*cos(lh^2/(mic*sqrt(mic^2*cos(αd)^2 + (mic*cos(αd)*tan(αd) + sqrt(-mic^2*cos(αd)^2/2 + sqrt(4*lh^4 + mic^4*cos(αd)^4)/2))^2)) + αd + αm)/(Wog*sin(lh^2/(mic*sqrt(mic^2*cos(αd)^2 + (mic*cos(αd)*tan(αd) + sqrt(-mic^2*cos(αd)^2/2 + sqrt(4*lh^4 + mic^4*cos(αd)^4)/2))^2)) + αd + αm)) + tan(αd)^2 + 1
                             α=αd-f/fprime
                                               
                             if isnan(α)==true
                                               L=αd
                                               eps=abs((L-αd)/L)
                                               ITNumb=ITNumb+0
                             else
                                               L=α  
                                               eps=abs((L-αd)/L)
                                               ITNumb=ITNumb+1
                             end                              
                 end
                                           
            end # of while loop eps>EPS
                     
            if isnan(L)==true
                      nan=nan+1
                      res=["WARNING: NaN No.$nan"]
                      println(res)
            end
         
            ######## LOADER ########                            
            if 0<=L<pi/2
               k=k+1
            if k==1
               L1=L
            elseif k==2
            if abs(L-L1)*57.3>0.1
                L2=L          
            end
            else
            if abs(L-L1)*57.3>0.1
            if L2==100
               L2=L
            else
            if abs(L-L2)*57.3>0.1                    
               L3=L
            end
            end
            end
            end
            end                                
            ########################                                                                  

      end # of for loop αd0 in [0, 10, ... 89.9]
                                                               
      for l in [L1, L2, L3]
      if 0<=l<pi/2
                                           
                 ################ THE FINISHER #######################                          
                  Mat=(V*cos(l)+Vt)/SOS
                  Mr=(0.75*Vt+V*cos(l))/SOS                          
                  Mtip=Vt/SOS
                  mi=mic*cos(l)
                  miz=mi*tan(l)
                  le=sqrt(sqrt(mi^4+4*lh^4)-mi^2)/sqrt(2)
                  li=(lh^2)/sqrt((le+miz)^2+mi^2)
                  αdw=li/mic
                  αf=(l)+(αdw)+(αm)  
                                           
                  if Mat<=0.9
                            al=alo/sqrt(1-(0.75*Mtip)^2)
                  elseif Mat<1.1
                            al=57.3*0.055*(0.26*(Mr)^2)^(-1/3)            
                  else
                            al=0.001
                  end  
                                           
                  PD=PDo*(A+B*(sin(αf))^n)
                  T=Wog*cos(l)+PD*sin(l)
                  Ct=T/c
                  Clav=6*Ct/sigma/(1+1.5*mi^2)
                  αl=Clav/al+αn
                                           
                  if Mr<=0.8
                            cd=δo+δ1*αl+δ2*αl^2+δ3*αl^3
                  elseif Mr<=1
                            cd=δo+δ1*αl+δ2*αl^2+δ3*αl^3+0.4*(Mr-0.8)
                  else
                            cd=1
                  end
                                           
                  H=(3.65/8)*pi*ro*sigma*cd*Vt*V*R^2*cos(l)
                  AoADGREE=l*180/pi
                  AoA=round(AoADGREE, digits=3)
                  μ=round(mi, digits=3)
                  CT=round(Ct, digits=8)          
                  #cf1=T*sin(L)-H*cos(L)-PD
                  #cf2=T*cos(L)+H*sin(L)-Wog
                  data_in=[μ, AoA, CT]
                  Coeff1=data_in[1]
                  Coeff2=data_in[2]
                  Coeff3=data_in[3]
                                           
                 ############# END FINISHER ######################
                                           
                 ############# THE PUSHER ########################
                    
                  k1=k1+1                              
                  ###############################                              
                  
                  ################################                              
                  push!(v1, [Coeff1, Coeff2, Coeff3])
                  n1=n1+1 
                  
                 ############## THE END of PUSHER ################
     end
     end                                                                                  
                                end# for Velocity, V
                            end
                        end
                    end
                end
            end
        end
    end
res=[v1, n1, k1, Jack1]
return(res)
end
```

```julia
# Vektor V NE mora biti prethodno sortiran, i NULA NULA NULA nije izbacena
function Worker(v,n)       
       Worker1=[[0.0, 0.0]]
       Worker2=[[0.0, 0.0]]
       Workers2=[[0.0, 0.0]]
       ix_drop = 1:1
       deleteat!(v,ix_drop)
       v=v[sortperm(v[:, 1]), :]    
       push!(Worker1, [v[1][1], v[1][2]])
       push!(Workers2, [v[1][1], v[1][3]])
       p1w1=1
       p1ws2=1
for i=1:n-2
    if v[i+2][1]==v[i][1]
    else
       push!(Worker1, [v[i+1][1], v[i+1][2]])
       p1w1=p1w1+1 
    end
    end
       push!(Worker1, [v[n][1], v[n][2]])
       p1w1=p1w1+1
for i=1:n-2
    if v[i+2][2]==v[i][2]
    else
       push!(Workers2, [v[i+1][1], v[i+1][3]])
       p1ws2=p1ws2+1
    end
    end
       push!(Workers2, [v[n][1], v[n][3]])
       p1ws2=p1ws2+1
ix_drop = 1:1
deleteat!(Worker1,ix_drop)
deleteat!(Workers2,ix_drop)
Workers2=Workers2[sortperm(Workers2[:, 1]), :]
push!(Worker2, [Workers2[1][1], Workers2[1][2]])
p1w2=1
     for i=1:p1ws2-2
         if Workers2[i+2][1]==Workers2[i][1]
         else
         push!(Worker2, [Workers2[i+1][1], Workers2[i+1][2]])
         p1w2=p1w2+1
    end
    end
       push!(Worker2, [Workers2[p1ws2][1], Workers2[p1ws2][2]])
       p1w2=p1w2+1

ix_drop = 1:1
deleteat!(Worker2,ix_drop)
return([Worker1, Worker2, p1w1, p1w2])
end
```

```julia
function Mixer(WA11,WA12,WA21,WA22,WA31,WA32,WA41,WA42,WA51,WA52,WA61,WA62,WA71,WA72,WA81,WA82,mA11,mA12,mA21,mA22,mA31,mA32,mA41,mA42,mA51,mA52,mA61,mA62,mA71,mA72,mA81,mA82)
    
for i=1:mA21
    push!(WA11,WA21[i])    
end  
for i=1:mA31
    push!(WA11,WA31[i])  
end 
for i=1:mA41
    push!(WA11,WA41[i])   
end
for i=1:mA51
    push!(WA11,WA51[i])    
end 
for i=1:mA61
    push!(WA11,WA61[i])    
end
for i=1:mA71
    push!(WA11,WA71[i])    
end 
for i=1:mA81
    push!(WA11,WA81[i])    
end 
    
for i=1:mA22   
    push!(WA12,WA22[i])
end  
for i=1:mA32  
    push!(WA12,WA32[i])
end 
for i=1:mA42 
    push!(WA12,WA42[i])
end    
for i=1:mA52 
    push!(WA12,WA52[i])
end  
for i=1:mA62 
    push!(WA12,WA62[i])
end  
for i=1:mA72 
    push!(WA12,WA72[i])
end  
for i=1:mA82 
    push!(WA12,WA82[i])
end     
m1=mA11+mA21+mA31+mA41+mA51+mA61+mA71+mA81
m2=mA12+mA22+mA32+mA42+mA52+mA62+mA72+mA82    
    
W1=WA11[sortperm(WA11[:, 1]), :] 
W2=WA12[sortperm(WA12[:, 1]), :] 
final_min_max_1=[[0.0,0.0]]
final_min_max_2=[[0.0,0.0]]
final_min_1=[[0.0,0.0]]
final_min_2=[[0.0,0.0]]
final_max_1=[[0.0,0.0]]
final_max_2=[[0.0,0.0]]
push!(final_min_max_1,[W1[1][1],W1[1][2]])
push!(final_min_max_2,[W2[1][1],W2[1][2]])
mm1=1
mm2=1
for i=1:m1-2
    if W1[i+2][1]==W1[i][1]
    else 
    push!(final_min_max_1,[W1[i+1][1],W1[i+1][2]])
    mm1=mm1+1
end
end    
    push!(final_min_max_1,[W1[m1][1],W1[m1][2]])
    mm1=mm1+1

for i=1:m2-2
    if W2[i+2][1]==W2[i][1]
    else 
    push!(final_min_max_2,W2[i+1])
    mm2=mm2+1    
end
end
    push!(final_min_max_2,W2[m2])
    mm2=mm2+1
ix_drop = 1:1
deleteat!(final_min_max_1,ix_drop)        
deleteat!(final_min_max_2,ix_drop) 

Nomin1=0
Nomin2=0
Nomax1=0
Nomax2=0
    
if iseven(mm1)==true
   mmin1=mm1-1
else
   mmin1=mm1
end
    
for i=1:2:mmin1
    push!(final_min_1,final_min_max_1[i]) 
    Nomin1=Nomin1+1    
end
    
if iseven(mm2)==true
   mmin2=mm2-1
else
   mmin2=mm2
end 
    
for i=1:2:mmin2
    push!(final_min_2,final_min_max_2[i])
    Nomin2=Nomin2+1
end
    
if iseven(mm1)==true
   mmax1=mm1
else
   mmax1=mm1-1
end 
    
for i=2:2:mmax1
    push!(final_max_1,final_min_max_1[i])
    Nomax1=Nomax1+1
end
    
if iseven(mm2)==true
   mmax2=mm2
else
   mmax2=mm2-1
end  
    
for i=2:2:mmax2
    push!(final_max_2,final_min_max_2[i])
    Nomax2=Nomax2+1    
end

ix_drop= 1:1 
deleteat!(final_min_1,ix_drop)     
deleteat!(final_min_2,ix_drop) 
deleteat!(final_max_1,ix_drop)        
deleteat!(final_max_2,ix_drop) 
res=[final_min_1, final_max_1, final_min_2, final_max_2]
mimin1=[0.0]
mimax1=[0.0]    
mimin2=[0.0]
mimax2=[0.0]     
min1=[0.0]
max1=[0.0]
min2=[0.0]
max2=[0.0]
a=res[1]
b=res[2]
c=res[3]
d=res[4]

for i=1:Nomin1
push!(min1,a[i][2])
push!(mimin1,a[i][1])
end

for i=1:Nomax1
push!(max1,b[i][2])
push!(mimax1,a[i][1])
end
  
for i=1:Nomin2
push!(min2,c[i][2])
push!(mimin2,c[i][1])
end
      
for i=1:Nomax2
push!(max2,d[i][2])
push!(mimax2,c[i][1])
end              
             
ix_drop= 1:1      
deleteat!(mimin1,ix_drop) 
deleteat!(mimin2,ix_drop)
deleteat!(mimax1,ix_drop) 
deleteat!(mimax2,ix_drop)
deleteat!(min1,ix_drop) 
deleteat!(max1,ix_drop) 
deleteat!(min2,ix_drop) 
deleteat!(max2,ix_drop) 
result=[Nomin1, mimin1, min1, Nomax1, mimax1, max1, Nomin2, mimin2, min2, Nomax2, mimax2, max2]   
return(result)
end
```
